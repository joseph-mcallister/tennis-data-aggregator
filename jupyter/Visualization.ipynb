```python
import pandas
import plotly
import plotly.express as px
```

```python
players = pandas.read_csv('../data/Players.csv')
utr = pandas.read_csv('../data/UTREntries.csv')
```

```python
merged = players.merge(utr, how='left', on='utrProfileId')
merged = merged.drop_duplicates(subset='utrProfileId', keep="first")
```

```python
merged
```

```python
fig = px.scatter_geo(merged, 
                     lat="latitude", 
                     lon="longitude",
                     opacity=0.3,
                     title="High Likelihood D1 Tennis Players"
                     )
fig.show()
```
